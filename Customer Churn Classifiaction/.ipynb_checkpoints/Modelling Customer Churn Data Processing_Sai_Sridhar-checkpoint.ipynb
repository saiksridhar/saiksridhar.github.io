### EDA (From Course 1)

```python
# Define list of categorical variables
category_variables = (
    ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember', 'NumOfProducts', 'Complain', 'Satisfaction Score', 'Card Type']
)
```

```python
def cat_exit_viz_gen(df, category_list, row_id_variable, response_variable):
    '''Generates stacked barcharts for overall category splits for each element in category_list
    and category wise barcharts split by response_variable'''
    
    fig_cols = 4
    fig_rows = int(np.ceil(len(category_list)/4))
    fig, axs = plt.subplots(fig_rows, fig_cols, figsize = (15, 10))
    
    row_count = 0
    for row_num in range(fig_rows):
        for col_num in range(fig_cols): 
            
            if row_count + col_num >= len(category_list):
                axs[row_num, col_num].axis('off')
                continue
            
            cat = category_list[row_count + col_num]
            category_exit_group_data = df.groupby([cat, response_variable])[row_id_variable].count()
            category_group_data = df.groupby([cat])[row_id_variable].count()
            category_exit_group_summ = (category_exit_group_data.div(category_group_data, level = cat) * 100).unstack()
            
            # Plot bar chart to view exited percentage by category
            ax_plot = category_exit_group_summ.plot(
                kind = 'barh',
                stacked = True,
                xlabel = 'Percentage of Customers',
                ylabel = '',
                ax = axs[row_num, col_num]
            )
            
            for container in ax_plot.containers:
                ax_plot.bar_label(container, fmt = '%.1f', label_type='center')
            
            axs[row_num, col_num].set_title(f'Exits by {cat}')
        
        row_count += fig_cols
            
    return None
```

```python
cat_exit_viz_gen(base_df, category_variables, 'CustomerId', 'Exited')
```

Exits by complaints are extremely high. Essentially, the data is suggesting that if a customer has registered a complaint, there is a 99.5% chance that the customer will subsequently churn. 

There could be an issue of data leakage here which will need to be explored further. It is possible complaints are primarily picked up as a cusotmer is already in the churn process. This is something that will need to be clarified by the business.

```python
# For continuous variables:
# - Generate violin plots split by Exited flag
continuous_variables = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary', 'Point Earned']
```

```python
def cont_exit_viz_gen(df, continuous_variable_list, response_variable):
    '''Generates violin plots for overall data distibution for each variable in continuous_variable_list
    further split by the response_variable'''
    
    fig_cols = 4
    fig_rows = int(np.ceil(len(continuous_variable_list)/4))
    fig, axs = plt.subplots(fig_rows, fig_cols, figsize = (15, 10))
    
    row_count = 0
    for row_num in range(fig_rows):
        for col_num in range(fig_cols): 
            
            if row_count + col_num >= len(continuous_variable_list):
                axs[row_num, col_num].axis('off')
                continue
            
            col = continuous_variable_list[row_count + col_num]
            
            # Plot violin plot to view exited percentage by category
            ax_plot = sns.violinplot(
                x = response_variable, 
                y = col, 
                data = df, 
                orient = 'v', 
                width = 0.7,
                ax = axs[row_num, col_num]
            ) 
            
            
            axs[row_num, col_num].set_title(f'Exits by {col}')
        
        row_count += fig_cols
            
    return None
```

```python
cont_exit_viz_gen(base_df, continuous_variables, "Exited")
```

Distribution of customers who exited are quite similar for continuous variables. However, those who exited were overall older had slightly higher balances and also lower credit scores. These fields of CreditScore, Balance and Age may have reasonable predictive power for the model.

### Data Preperation

```python
# Split data into test and train sets
X_train, X_test, y_train, y_test = train_test_split(
    base_df.drop(columns = ['Exited']), 
    base_df[['Exited']], 
    test_size = 0.2,
    random_state = 23)
```

Data splitting is done at the initial stage to ensure no data leakage occurs as a result of data processing. Data processing pipelines will be developed for the train set which will then be used to transform the test set.

```python
set_config(transform_output = 'pandas')
```

```python
# Prep list of columns to drop for ML models
redundant_cols = ['RowNumber', 'CustomerId', 'Surname']

def redundant_cols_position_gen(df, cols):
    '''Generates a list of column positions to drop from a n-dimensional array'''
    redundant_cols_position = []
    base_df_cols = df.columns.to_list()

    for col in redundant_cols:
        redundant_cols_position.append(base_df_cols.index(col))
    
    return redundant_cols_position
```

```python
redundant_cols_position = redundant_cols_position_gen(base_df, redundant_cols)
```

```python
# Create custom data transformers
class ColDrop(BaseEstimator, TransformerMixin):
    '''Drops specified columns from a given array'''
    
    def __init__(self, col_list):
        
        assert type(col_list) == list, 'Columns to drop need to be specified as a list'
        
        self.col_list = col_list
    
    def fit(self, X, y = None):
        return self
    
    
    def transform(self, X, y = None):
        X = X.copy()
        
        df_cols = X.columns.to_list()
        
        cols_to_drop = [x for x in self.col_list if x in df_cols]
        
        X.drop(columns = cols_to_drop, inplace = True)
        
        return X
```

```python
transform_category_variables = category_variables.copy()
transform_category_variables.remove('NumOfProducts')
```

```python
# Create ColumnTransformer class
col_preprocessing = ColumnTransformer(transformers = [
    ('ColDrop', ColDrop(redundant_cols), redundant_cols),
    ('ohe', OneHotEncoder(drop = 'first', handle_unknown = 'ignore', sparse_output = False), transform_category_variables)
],
                                      verbose_feature_names_out=False,
                                      remainder = 'passthrough'
)
```

```python
# Create initial classifier pipelines
# Logistic Regression Classifier 
log_reg_clf = Pipeline(steps = [
    ('col_preprocessing', col_preprocessing),
    ('scaler', StandardScaler()),
    ('log_reg', LogisticRegression())
])

# Random Forest Classifier
rf_clf = Pipeline(steps = [
    ('col_preprocessing', col_preprocessing),
    ('rf_clf', RandomForestClassifier())
])


# GBM Classifier
gbm_clf = Pipeline(steps = [
    ('col_preprocessing', col_preprocessing),
    ('gbm_clf', GradientBoostingClassifier())
])
```

```python
# Fit untuned logistic regression model
log_reg_clf.fit(X_train, y_train.values.ravel())
```

```python
# Fit untuned random forest model
rf_clf.fit(X_train, y_train.values.ravel())
```

```python
# Fit untuned GBM model
gbm_clf.fit(X_train, y_train.values.ravel())
```

```python
y_pred_log_reg = log_reg_clf.predict(X_test)
y_pred_proba_log_reg = log_reg_clf.predict_proba(X_test)
```

```python
y_pred_rf = rf_clf.predict(X_test)
y_pred_proba_rf = rf_clf.predict_proba(X_test)
```

```python
y_pred_gbm = gbm_clf.predict(X_test)
y_pred_proba_gbm = gbm_clf.predict_proba(X_test)
```

```python
def classification_performance_report(model_name, y_test, y_preds_proba, y_preds):
    '''Displays the performance of a binary classifier output'''
    
    # Get predicted probabilities 
    y_scores = y_preds_proba[:,1]

    # Calculate and visualize ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_scores, pos_label=0)
    plt.plot(tpr, fpr, label = 'ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {model_name} model')
    plt.legend()
    plt.show()
    
    # Calculate AUC
    auc = roc_auc_score(y_test, y_scores)
    print(f'AUC: {auc}\n')

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_preds)
    print(f'Confusion Matrix:\n{cm}\n')

    # Other metrics
    precision = precision_score(y_test, y_preds)
    recall = recall_score(y_test, y_preds)
    accuracy = accuracy_score(y_test, y_preds)
    print(f'Precision: {precision}, Recall: {recall}, Accuracy: {accuracy}')
    
    return None
```

```python
# Classification performance report for Linear Regression
classification_performance_report('Linear Regression', y_test, y_pred_proba_log_reg, y_pred_log_reg)
```

```python
# Classification performance report for Random Forest
classification_performance_report('Random Forest', y_test, y_pred_proba_rf, y_pred_rf)
```

```python
# Classification performance report for GBM
classification_performance_report('GBM', y_test, y_pred_proba_gbm, y_pred_gbm)
```

_Insights_

All models generate perfect or near perfect predictions which seems unrealistic in the real world in terms of predicting customer churn.

Since the data was split into train and test sets prior to any data processing, there is no room for data pre-processing leakage.

However, there could be information present within the predictor variables that may not be present when making predictions in the real world. There is a possibility of leakage in labels.

For this, we would need to investigate the best features for each model and study them further.

```python
# retrieving the RF Classifier from the model pipeline
def feat_importance(model_pipeline, model_name = ''):
    '''Generates a barchart for a given model to depict the models feature importance'''

    clf = model_pipeline[-1]
    # making a pandas dataframe
    if type(clf) == LogisticRegression:
        break
    else:
        data = list(zip(clf.feature_names_in_, clf.feature_importances_))
        df_importances = pd.DataFrame(data, columns=['Feature', 'Importance']).sort_values(by='Importance', ascending=False)
        df_importances.sort_values(['Importance']).plot(x='Feature', y='Importance', kind='barh', figsize=(10, 6))
        plt.title(f'Feature Importance for {model_name} model')
        plt.show()
    return None
```
