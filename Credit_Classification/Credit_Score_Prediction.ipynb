# Credit Score Predictions

The Credit Score classification dataset explored includes a number of customer features over the first 8 months of a given year. These include key customer features such as the number of products, cashflow, arrears, total repayments and the like which should allow a good view of customer behaviour and hence support the prediction of the next 4 months of customer credit scores.

Credit Scoring enables credit decisioning and pricing (such as interest rates) to support an optimised portfolio for the institution in line with profitablility targets and risk apetite. This is vital to the survival and success of any financial institution which issues lending products.

While the data is a time series per customer, credit scores would be typically be determined at a point in time. Average of previous performance metrics would be considered however. The approach in this notebook is to experiment with various Neural Network configurations in order to develop predictive algorithm for the financial institution.

The primary challenges dealt with in this classification process are data cleansing, imputation and model development. Custom imputers have been developed in this Notebook along with testing of the imputed features during the model training phase. Additionally, various configurations and approaches are taken to develop a predictive neural network model.

```python
# Create number cleaner and formatter to pre-process fields that should be numeric
num_cols = [
    'Age',
    'Annual_Income',
    'Monthly_Inhand_Salary',
    'Num_Bank_Accounts',
    'Num_Credit_Card',
    'Interest_Rate',
    'Num_of_Loan',
    'Delay_from_due_date',
    'Num_of_Delayed_Payment',
    'Changed_Credit_Limit',
    'Num_Credit_Inquiries',
    'Outstanding_Debt',
    'Credit_Utilization_Ratio',
    'Total_EMI_per_month',
    'Amount_invested_monthly',
    'Monthly_Balance'
]
    
class NumberCleanser(BaseEstimator, TransformerMixin):
    '''Given a list of columns that should be numeric, this class cleans and transforms each column to a float. This process
    would need to be run prior to running the dataset through a column transformer as it produces a correctly formatted dataset that can 
    then be transformed where required.'''
    
    def __init__(self, transform_cols):        
        self.transform_cols = transform_cols
    
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = X.copy()
        
        for col in self.transform_cols:
            X[col] = X[col].astype('str').str.extract(r"(\-?\d*\.?\d*)").replace('', None).astype('float64')
        
        return X
```

```python
# Create transformer to drop redundant columns post transformations
class ColDrop(BaseEstimator, TransformerMixin):
    '''Drops specified columns from a given array'''
    
    def __init__(self, col_list):        
        self.col_list = col_list
    
    def fit(self, X, y = None):
        return self
    
    
    def transform(self, X, y = None):
        X = X.copy()
        
        col_count = X.shape[1]
        cols_to_keep = list(range(0, col_count))
        cols_to_remove = self.col_list

        cols_to_keep = [x for x in cols_to_keep if x not in cols_to_remove]
        X = X[:, cols_to_keep]
        
        return X
        
        return X
```

```python
# Create custom transformers
class CustomColumnDrop(BaseEstimator, TransformerMixin):
    '''
    Customized transformer which drops predefined columns from dataset
    '''
    
    def __init__(self, col_list):
        assert type(col_list) == list, 'Columns to drop need to be specified as a list (even if single column)'
        
        self.col_list = col_list
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = X.copy()
        
        col_count = X_train.shape[1]
        cols_to_keep = list(range(0, col_count))
        cols_to_remove = self.col_list

        cols_to_keep = [x for x in cols_to_keep if x not in cols_to_remove]
        X = X[:, cols_to_keep]
        
        return X

    
class CSVSplitOHE(BaseEstimator, TransformerMixin):
    '''
    Customized Transformer which splits categorical columns with multiple comma separated categories 
    into One Hot Encoded output. This returns a pandas dataframe with added columns representing the 
    one hot encoded categories determined within the processed column.
    '''
    def __init__(self, colname, replace_with_empty_string = None, replace_with_space = None):
        assert type(colname) == str, 'Input column for transformation must be expressed as a string'
        assert (type(replace_with_empty_string) == list) or (replace_with_blank is None), 'Input must be a list of strings (even if input is a single item)'
        assert (type(replace_with_space) == list) or (replace_with_space is None), 'Input must be a list of strings (even if input is a single item)'
        
        self.colname = colname
        self.replace_with_empty_string = replace_with_empty_string
        self.replace_with_space = replace_with_space
        
    def fit(self, X, y = None):
        category_types = []
        for cat_csv in X[self.colname].fillna('').str.split(','):
            for cat in cat_csv:
                clean_cat = cat
                if self.replace_with_empty_string is not None:
                    for item in self.replace_with_empty_string:
                        clean_cat = clean_cat.replace(item, '')
                if self.replace_with_space is not None:
                    for item in self.replace_with_space:
                        clean_cat = clean_cat.replace(item, ' ')
                category_types.append(clean_cat.strip())
        category_types_dist = list(set(category_types))
        category_types_dist.remove('')
        
        self.category_types_list = category_types_dist
        
        return self
        
    def transform(self, X, y = None):
        check_is_fitted(self, 'category_types_list')
        
        X = X.copy()
        
        for cat in self.category_types_list:
            col_name = cat.replace(' ', '_') + '_flag'
            X[col_name] = X[self.colname].str.contains(cat).map({True: 1, False: 0})
            X[col_name].fillna(0, inplace = True)
        X.drop(columns = [self.colname], inplace = True)
        
        return X

    
class GroupLevelMetricImputer(BaseEstimator, TransformerMixin):
    '''
    Customized transformer that calculates specified metric at a given grouping level. Transformation then replaces NaN values with
    fitted imputer values at group level.
    '''
    
    def __init__(self, group_level, impute_colname, imputer_metric):
        assert type(group_level) == list, 'Grouping columns need to be specified as a list'
        assert type(impute_colname) == str, 'Transformation column needs to be specified as a string variable'
        assert ['mean', 'max', 'min', 'zero'].__contains__(imputer_metric), 'imputer metric is not in accepted options which are mean, max, min or zero'
        
        self.group_level = group_level
        self.impute_colname = impute_colname
        self.imputer_metric = imputer_metric
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        
        X = X.copy()
        
        X[self.impute_colname] = X[self.impute_colname].astype('str').str.extract(r"([\d\.\-]+)").astype('float64')
        
        if self.imputer_metric == 'mean':
            imputer_mapper = X.groupby(self.group_level)[self.impute_colname].mean().reset_index()
        
        elif self.imputer_metric == 'max':
            imputer_mapper = X.groupby(self.group_level)[self.impute_colname].max().reset_index()

        elif self.imputer_metric == 'min':
            imputer_mapper = X.groupby(self.group_level)[self.impute_colname].min().reset_index()
            
        else:
            imputer_mapper = None
        
        if imputer_mapper is not None:
            imputer_mapper.rename({self.impute_colname: '_impute_map_column'}, axis = 1, inplace = True)
            X = X.merge(imputer_mapper, on = self.group_level, how = 'left')
            X[self.impute_colname] = X[self.impute_colname].combine_first(X['_impute_map_column'])
            X.drop(columns = ['_impute_map_column'], inplace = True)
        
        else:
            X[self.impute_colname] = X[self.impute_colname].fillna(0)
        
        X.drop(columns = self.group_level, inplace = True)
        
        return X

    
class GroupLevelMonthImputer(BaseEstimator, TransformerMixin):
    '''
    Customized transformer which converts periods expressed as years and months to months only. Populates null values by deducing months in reference to first populated value
    for the group
    '''
    
    def __init__(self, group_level, transform_colname, month_colname):
        assert type(group_level) == list, 'Grouping columns need to be specified as a list'
        assert type(transform_colname) == str, 'Transformation column needs to be specified as a string variable'
        assert type(month_colname) == str, 'Month column needs to be specified as a string variable'
        
        self.group_level = group_level
        self.transform_colname = transform_colname
        self.month_colname = month_colname
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        
        X = X.copy()
        
        X[self.transform_colname] = X[self.transform_colname].str.findall('\d+')
        X[self.transform_colname] = X[self.transform_colname].apply(lambda x: (int(x[0]) * 12) + int(x[1]) if type(x) == list else np.nan)
        
        months_list = list(calendar.month_name)
        
        X['_Enumerated_' + self.month_colname] = X[self.month_colname].apply(lambda x: months_list.index(x))
        
        X['_Min_Ranked_' + self.transform_colname] = X[X[self.transform_colname].notna()].groupby(self.group_level)['_Enumerated_' + self.month_colname].rank('min')
        
        select_cols = []
        
        for col in self.group_level:
            select_cols.append(col)
        
        select_cols += [self.transform_colname, '_Enumerated_' + self.month_colname]
        
        min_month_value_df = X[X[('_Min_Ranked_' + self.transform_colname)] == 1.0][select_cols].rename(columns = {
            self.transform_colname: '_Min_Value_' + self.transform_colname,
            '_Enumerated_' + self.month_colname: '_Min_Enumerated_' + self.month_colname
        })
        
        X = X.merge(
            min_month_value_df,
            on = self.group_level,
            how = 'left'
        )
        
        X['_Adjustment_' + self.transform_colname] = X['_Enumerated_' + self.month_colname] - X['_Min_Enumerated_' + self.month_colname]
        
        X[self.transform_colname] = X.apply(lambda x: x['_Min_Value_' + self.transform_colname] + x['_Adjustment_' + self.transform_colname]  if math.isnan(x[self.transform_colname]) else x[self.transform_colname], axis = 1)
        
        X.drop(columns = ['_Enumerated_' + self.month_colname, 
                          '_Min_Ranked_' + self.transform_colname,
                          '_Min_Value_' + self.transform_colname,
                          '_Min_Enumerated_' + self.month_colname,
                          '_Adjustment_' + self.transform_colname,
                          self.month_colname
                         ] + self.group_level, inplace = True)
        
        return X

    
class GroupLevelOrdinalEncoder(BaseEstimator, TransformerMixin):
    '''
    Customized ordinal encoder which fits a pre-defined ordered list, enumerates transformation column and fills blanks with average
    of enumerated value at a given group level
    '''
    def __init__(self, group_level, transform_column, ordered_cats):
        assert type(group_level) == list, 'Grouping columns need to be specified as a list'
        assert type(transform_column) == str, 'Transformation column needs to be specified as a string variable'
        assert type(ordered_cats) == list, 'Categories must be provided as an ordered list (if order is relevant)'
        
        self.group_level = group_level
        self.transform_column = transform_column
        self.ordered_cats = ordered_cats
        
    def fit(self, X, y = None):
        
        self.ordered_cats_fitted = {}
        for i in self.ordered_cats:
            self.ordered_cats_fitted[i] = self.ordered_cats.index(i) + 1
        
        return self
    
    def transform(self, X, y = None):
        
        check_is_fitted(self, 'ordered_cats_fitted')
        
        X = X.copy()
        
        X[self.transform_column] = X[self.transform_column].map(self.ordered_cats_fitted)
        
        group_summ_df = X.groupby(self.group_level)[self.transform_column].mean().reset_index()
        
        group_summ_df.rename({self.transform_column: '_impute_col_summ'}, axis = 1, inplace = True)
        
        X = X.merge(
            group_summ_df,
            on = self.group_level,
            how = 'left'
        )
          
        X[self.transform_column] = X[self.transform_column].combine_first(X['_impute_col_summ'])
        
        X.drop(columns = self.group_level + ['_impute_col_summ'], inplace = True)
        
        return X

class FillNaTranformer(BaseEstimator, TransformerMixin):
    '''
    Fills null values with given value
    '''
    def __init__(self, transform_cols, fill_with = 0):
        assert type(transform_cols) == list, 'Transformation columns need to be specified within a list'
        self.transform_cols = transform_cols
        self.fill_with = fill_with
        
        return None
    
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        X = X.copy()
        for col in self.transform_cols:
            X[col] = X[col].fillna(self.fill_with)
        
        return X
    
    
class NonTranformer(BaseEstimator, TransformerMixin):
    '''
    Returns columns unprocessed. This is to account for columns that were passed to other transformers for functional
    purpose but were not transformed in any way themselves. Also, it positions unwanted columns at the start of the
    output array to make for easy removal.
    '''
    def __init__(self):
        return None
    
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        return X
```

```python
# Create ColumnTransformer class
col_preprocessing = ColumnTransformer(transformers = [
    ('null_transform', NonTranformer(), ['ID', 'Customer_ID', 'Month', 'Name', 'SSN']),
    ('mnth_saly_imp', GroupLevelMetricImputer(['Customer_ID'], 'Monthly_Inhand_Salary', 'mean'), ['Customer_ID', 'Monthly_Inhand_Salary']),
    ('type_loan_imp', CSVSplitOHE('Type_of_Loan', ['Not Specified', 'and'], ['-']), ['Type_of_Loan']),
    ('num_dely_pymt_imp', GroupLevelMetricImputer(['Customer_ID'], 'Num_of_Delayed_Payment', 'mean'), ['Customer_ID', 'Num_of_Delayed_Payment']),
    ('num_crdt_enqs_imp', GroupLevelMetricImputer(['Customer_ID'], 'Num_Credit_Inquiries', 'mean'), ['Customer_ID', 'Num_Credit_Inquiries']),
    ('hist_age_enum_enc', GroupLevelMonthImputer(['Customer_ID'], 'Credit_History_Age', 'Month'), ['Customer_ID', 'Credit_History_Age', 'Month']),
    ('amnt_invs_mnth', GroupLevelMetricImputer(['Customer_ID'], 'Amount_invested_monthly', 'mean'), ['Customer_ID', 'Amount_invested_monthly']),
    ('mnth_baln_imp', GroupLevelMetricImputer(['Customer_ID'], 'Monthly_Balance', 'mean'), ['Customer_ID', 'Monthly_Balance']),
    ('pymt_behv', GroupLevelOrdinalEncoder(['Customer_ID'], 'Payment_Behaviour', [
        'Low_spent_Small_value_payments',
        'Low_spent_Medium_value_payments',
        'Low_spent_Large_value_payments',    
        'High_spent_Small_value_payments',
        'High_spent_Medium_value_payments',
        'High_spent_Large_value_payments']), ['Customer_ID', 'Payment_Behaviour']),
    ('crdt_mix_enc', GroupLevelOrdinalEncoder(['Customer_ID'], 'Credit_Mix', ['Good', 'Standard', 'Bad']), ['Customer_ID', 'Credit_Mix']),
    ('pymt_min_amt_enc', GroupLevelOrdinalEncoder(['Customer_ID'], 'Payment_of_Min_Amount', ['Yes', 'No', 'NM']), ['Customer_ID', 'Payment_of_Min_Amount']),
    ('ocpn_enc', OrdinalEncoder(), ['Occupation']),
    ('FillNaTranformer', FillNaTranformer(['Changed_Credit_Limit'], 0), ['Changed_Credit_Limit'])
],
                             remainder = 'passthrough'
)
```

```python
# Create data processing pipeline - transformation can be used for models that require standardization as well as
# those that do not
data_process_pipeline = Pipeline(steps = [
    ('NumberCleanser', NumberCleanser(transform_cols = num_cols)),
    ('col_preprocessing', col_preprocessing),
    ('ColDrop', ColDrop(col_list = [0, 1, 2, 3, 4])),
    ('scaler', StandardScaler())
])
```

```python
label_enc = LabelEncoder()
```

## Models

```python
# Create neural network model builder
def build_model(n_hidden, n_neurons, input_shape, output_shape):
    
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape = input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation = 'relu'))
    model.add(keras.layers.Dense(output_shape))
    
    return model

nn_model = KerasClassifier(
    model = build_model,
    loss = 'sparse_categorical_crossentropy',
    optimizer = keras.optimizers.SGD(learning_rate = 0.01),
    epochs = 50, 
    batch_size = 1000, 
    verbose = 0,
    n_hidden = 1, 
    n_neurons = 30,
    input_shape = [29],
    output_shape = 3)
```

## Pipeline Creation

```python
# Label encoding
y_train_enc = label_enc.fit_transform(y_train.squeeze())
y_test_enc = label_enc.transform(y_test.squeeze())
```

```python
# Create basic XGBoost pipeline
clf_pipe_xgb = Pipeline(steps = [
    ('data_process_pipeline', data_process_pipeline),
    ('model', XGBClassifier())]
)
```

```python
# Create SVM pipeline
clf_pipe_svm = Pipeline(steps = [
    ('data_process_pipeline', data_process_pipeline),
    ('model', SVC(gamma = 'auto'))]
)
```

```python
# Create NN pipeline
# Due to batched approach to training NN model, data processing will not be built into the pipeline as 
# a complete dataset is required to impute missing data
clf_pipe_nn = Pipeline(steps = [
    ('model', nn_model)]
)
```
